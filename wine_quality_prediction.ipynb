```python
# K-Nearest Neighbous(KNN) with GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from xgboost import XGBClassifier
import optuna
from scipy.stats import randint

# Define parameter grid
knn_params = {'n_neighbors': [3, 5, 7, 9]}

# Create GridSearchCV
knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5, scoring='accuracy')
knn_grid.fit(X_train, y_train)

# Predict
knn_predict = knn_grid.predict(X_test)

# Evaluate
print("KNN")
print("Best Params:", knn_grid.best_params_)
print("Accuracy:", round(accuracy_score(y_test, knn_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, knn_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, knn_predict))
print("Classification Report:\n", classification_report(y_test, knn_predict))
```

```python
# Logistic Regression with GridSearch

from sklearn.linear_model import LogisticRegression

# Define parameter grid
logreg_params = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

# Create GridSearchCV
logreg_grid = GridSearchCV(LogisticRegression(max_iter=1000), logreg_params, cv=5, scoring='accuracy')
logreg_grid.fit(X_train, y_train)
# Predict
logreg_predict = logreg_grid.predict(X_test)

# Evaluate
print("Logistic Regression")
print("Best Parameters:", logreg_grid.best_params_)
print("Accuracy:", round(accuracy_score(y_test, logreg_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, logreg_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, logreg_predict))
print("Classification Report:\n", classification_report(y_test, logreg_predict))
```

```python
 # DecisionTree with GridSearchCV

from sklearn.tree import DecisionTreeClassifier

# Define parameter grid
tree_params = {'max_depth': [5, 10, 15], 'criterion': ['gini', 'entropy']}

# Create GridSearchCV
tree_grid = GridSearchCV(DecisionTreeClassifier(), tree_params, cv=5, scoring='accuracy')
tree_grid.fit(X_train, y_train)

# Predict
tree_predict = tree_grid.predict(X_test)

# Evaluate
print("Decision Tree")
print("Best Parameters:", tree_grid.best_params_)
print("Accuracy:", round(accuracy_score(y_test, tree_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, tree_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, tree_predict))
print("Classification Report:\n", classification_report(y_test, tree_predict))
```

```python
# RandomForest with RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define random search space
rf_params = {
    'n_estimators': randint(50, 150),
    'max_depth': randint(3, 15),
    'max_features': ['sqrt', 'log2']
}

# Create RandomizedSearchCV
rf_random = RandomizedSearchCV(RandomForestClassifier(random_state=42), rf_params,n_iter=10, cv=5, scoring='accuracy', random_state=42)
rf_random.fit(X_train, y_train)

# Predict
rf_predict = rf_random.predict(X_test)

# Evaluate
print("Random Forest")
print("Best Parameters:", rf_random.best_params_)
print("Accuracy:", round(accuracy_score(y_test, rf_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, rf_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_predict))
print("Classification Report:\n", classification_report(y_test, rf_predict))
```
